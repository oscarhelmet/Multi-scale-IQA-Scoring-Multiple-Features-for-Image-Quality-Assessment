```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches

# Read the CSV file
df = pd.read_csv('Dataset/Training_Set/data.csv', header = None)
labels = pd.read_csv('Dataset/Training_Set/label.csv', header = None)
labels = labels.astype(int)

# Create a 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Loop over each row
for i in range(len(df)):
  # Create x, z arrays using meshgrid
  x, z = np.meshgrid(np.arange(len(df.columns)), [i])

  # Get the y values
  y = df.iloc[i].values

  # Choose the color based on the label
  color = 'r' if labels.iloc[i, 0] == 0 else 'b'

  # Add the scatter plot for this row
  ax.scatter(x, z, y, color=color)

# Create a custom legend 
red_patch = mpatches.Patch(color='red', label='Label 0')
blue_patch = mpatches.Patch(color='blue', label='Label 1')
plt.legend(handles=[red_patch, blue_patch])

# Add titles and labels
ax.set_title('VGG Distances by Comparison')
ax.set_xlabel('Channel')
ax.set_ylabel('Comparison')
ax.set_zlabel('Value')

# Show the plot
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches

# Read the CSV file
df = pd.read_csv('Dataset/Training_Set/data.csv', header = None)
labels = pd.read_csv('Dataset/Training_Set/label.csv', header = None)
labels = labels.astype(int)

# Separate the dataframe based on the label
df_label0 = df[labels[0]==0]
df_label1 = df[labels[0]==1]

# Create a 3D plot for label 0
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Loop over each row in df_label0
for i in range(len(df_label0)):
    # Create x, z arrays using meshgrid
    x, z = np.meshgrid(np.arange(len(df_label0.columns)), [i])

    # Get the y values
    y = df_label0.iloc[i].values

    # Add the scatter plot for this row
    ax.scatter(x, z, y, color='r')

ax.set_title('VGG Distances by Comparison for Label 0')
ax.set_xlabel('Channel')
ax.set_ylabel('Comparison')
ax.set_zlabel('Value')

plt.show()

# Create a 3D plot for label 1
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Loop over each row in df_label1
for i in range(len(df_label1)):
    # Create x, z arrays using meshgrid
    x, z = np.meshgrid(np.arange(len(df_label1.columns)), [i])

    # Get the y values
    y = df_label1.iloc[i].values

    # Add the scatter plot for this row
    ax.scatter(x, z, y, color='b')

ax.set_title('VGG Distances by Comparison for Label 1')
ax.set_xlabel('Channel')
ax.set_ylabel('Comparison')
ax.set_zlabel('Value')

plt.show()
```
