```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as mcolors
import matplotlib.patches as mpatches

# Read the CSV file
df = pd.read_csv('Dataset/Training_Set/data.csv', header = None)
labels = pd.read_csv('Dataset/Training_Set/label.csv', header = None)
labels = labels.astype(int)

# Create a 3D plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Define a colormap
cmap = plt.get_cmap('tab10')  # 'tab10' has 10 distinct colors

# Loop over each row
for i in range(len(df)):
    # Create x, z arrays using meshgrid
    x, z = np.meshgrid(np.arange(len(df.columns)), [i])

    # Get the y values
    y = df.iloc[i].values

    # Choose the color based on the label, adjusting for zero-indexing
    color = cmap(labels.iloc[i, 0] - 1)

    # Add the scatter plot for this row
    ax.scatter(x, z, y, color=color)

# Create a custom legend
handles = [mpatches.Patch(color=cmap(i), label=f'Label {i+1}') for i in range(10)]
plt.legend(handles=handles, bbox_to_anchor=(1.05, 1), loc='upper left')

# Add titles and labels
ax.set_title('VGG Distances by Comparison')
ax.set_xlabel('Channel')
ax.set_ylabel('Comparison')
ax.set_zlabel('Value')

# Show the plot
plt.show()
```
